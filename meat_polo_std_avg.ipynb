```python
import os
import numpy as np
from PIL import Image

def process_images(base_folder, region=(75, 75, 175, 175)):
    categories = ['FRESH', 'HALF', 'SPOILT']
    colors = ['RED', 'GREEN', 'BLUE']
    parameters = ['D', 'Delta_output', 'P', 'P_delta_method1', 'P1', 'P2', 'P3', 'Ps', 'Psi_output', 'R_output']

    results = {}

    for category in categories:
        results[category] = {}
        for param in parameters:
            images = []
            for color in colors:
                folder_path = os.path.join(base_folder, f"{category}_{color}")
                file_path = os.path.join(folder_path, f"{param}.png")

                if os.path.exists(file_path):
                    image = Image.open(file_path).convert('L')  # Градации серого
                    image_array = np.array(image)
                    cropped = image_array[region[1]:region[3], region[0]:region[2]]  # Вырезаем регион
                    images.append(cropped)

            if images:
                avg_array = np.mean(images, axis=0)  # Усредняем по цветам
                results[category][param] = {
                    "mean": np.mean(avg_array),  # Среднее по центральной области
                    "std": np.std(avg_array)  # Стандартное отклонение
                }

    return results

# Пример использования
base_folder = "for dataset A"
result_values = process_images(base_folder)

for category, params in result_values.items():
    print(f"{category}:")
    for param, values in params.items():
        print(f"  {category}_{param} = {values['mean']:.2f}, std = {values['std']:.2f}")
```

```python
import os
import numpy as np
from PIL import Image

def process_images(base_folder, region=(75, 75, 175, 175)):
    categories = ['FRESH', 'HALF', 'SPOILT']
    colors = ['RED', 'GREEN', 'BLUE']
    parameters = ['D', 'Delta_output', 'P', 'P_delta_method1', 'P1', 'P2', 'P3', 'Ps', 'Psi_output', 'R_output']

    results = {}

    for category in categories:
        results[category] = {}
        for param in parameters:
            for color in colors:
                folder_path = os.path.join(base_folder, f"{category}_{color}")
                file_path = os.path.join(folder_path, f"{param}.png")

                if os.path.exists(file_path):
                    image = Image.open(file_path).convert('L')  # Градации серого
                    image_array = np.array(image)
                    cropped = image_array[region[1]:region[3], region[0]:region[2]]  # Вырезаем регион
                    mean_value = np.mean(cropped)  # Среднее значение
                    std_value = np.std(cropped)  # Стандартное отклонение
                    # Сохраняем значения для каждого цвета и параметра
                    results[category][f"{color}_{param}"] = {
                        "mean": mean_value,
                        "std": std_value
                    }

    return results

# Пример использования
base_folder = "for dataset A"
result_values = process_images(base_folder)

# Выводим результаты в нужном формате
for category, params in result_values.items():
    for param, values in params.items():
        print(f"{category}_{param} = {values['mean']:.2f}, std = {values['std']:.2f}")
```
