```python
import shutil
import os

def zip_folder(folder_path):
    """–ê—Ä—Ö–∏–≤–∏—Ä—É–µ—Ç –ø–∞–ø–∫—É –≤ ZIP —Å —Ç–µ–º –∂–µ –∏–º–µ–Ω–µ–º (–±–µ–∑ /content/)"""
    folder_name = os.path.basename(folder_path)  # –ü–æ–ª—É—á–∞–µ–º –∏–º—è –ø–∞–ø–∫–∏ –±–µ–∑ –ø—É—Ç–∏
    zip_file_name = f"{folder_name}.zip"  # –î–æ–±–∞–≤–ª—è–µ–º .zip

    try:
        shutil.make_archive(zip_file_name[:-4], 'zip', folder_path)  # –°–æ–∑–¥–∞—ë–º –∞—Ä—Ö–∏–≤
        print(f"‚úÖ –£—Å–ø–µ—à–Ω–æ –∞—Ä—Ö–∏–≤–∏—Ä–æ–≤–∞–Ω–æ: {folder_name} ‚Üí {zip_file_name}")
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –∞—Ä—Ö–∏–≤–∞—Ü–∏–∏: {e}")

# üîπ –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ
folder_to_zip1 = "/content/HALF_RED_M"  # –£–∫–∞–∂–∏ —Å–≤–æ—é –ø–∞–ø–∫—É
folder_to_zip2 = "/content/HALF_RED_P1P2P3"
zip_folder(folder_to_zip1)
zip_folder(folder_to_zip2)
```

```python
import os
import numpy as np
from PIL import Image

# –ü—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏
input_dir = "/content/SPOILT_BLUE"  # –ó–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –ø—É—Ç—å –∫ –≤–∞—à–µ–π –ø–∞–ø–∫–µ
output_dir = "HALF_RED_M"
# –°–ø–∏—Å–æ–∫ –¥–≤—É—Ö–±—É–∫–≤–µ–Ω–Ω—ã—Ö –∫–æ–¥–æ–≤
codes = [
    "HP", "PP", "VP", "QP", "RP", "LP", "LV", "RV", "VV", "PV", "HV", "QV",
    "QQ", "HQ", "PQ", "VQ", "RQ", "LQ", "LH", "RH", "HH", "QH", "PH", "VH",
    "RR", "LR", "HR", "QR", "PR", "VR", "VL", "PL", "HL", "QL", "RL", "LL"
]

# –°–ª–æ–≤–∞—Ä—å –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
images = {}

# –ó–∞–≥—Ä—É–∑–∫–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –∏–∑ –ø–∞–ø–∫–∏
for filename in os.listdir(input_dir):
    if filename.endswith(".png"):
        # –ò–∑–≤–ª–µ–∫–∞–µ–º –¥–≤—É—Ö–±—É–∫–≤–µ–Ω–Ω—ã–π –∫–æ–¥ –∏–∑ –Ω–∞–∑–≤–∞–Ω–∏—è —Ñ–∞–π–ª–∞
        code = filename.split(".")[1]  # –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ–º —Ñ–æ—Ä–º–∞—Ç "—á–∏—Å–ª–æ.–∫–æ–¥.png"
        if code in codes:
            file_path = os.path.join(input_dir, filename)
            img = Image.open(file_path)
            images[code] = np.array(img)
        else:
            print(f"–§–∞–π–ª {filename} —Å–æ–¥–µ—Ä–∂–∏—Ç –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω—ã–π –∫–æ–¥: {code}")

# –ü—Ä–æ–≤–µ—Ä–∫–∞, —á—Ç–æ –≤—Å–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –∑–∞–≥—Ä—É–∂–µ–Ω—ã
if len(images) != len(codes):
    print("–ù–µ –≤—Å–µ —Ñ–∞–π–ª—ã –±—ã–ª–∏ –∑–∞–≥—Ä—É–∂–µ–Ω—ã. –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –Ω–∞–ª–∏—á–∏–µ –≤—Å–µ—Ö —Ñ–∞–π–ª–æ–≤.")
else:
    # –°–æ–∑–¥–∞–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ M
    M = np.zeros((250, 250, 4, 4))

    # –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–π
    M[:, :, 0, 0] = np.clip((0.25 * images["HH"] + 0.25 * images["HV"] + 0.25 * images["VH"] + 0.25 * images["VV"]), 0, None) # M00
    M[:, :, 0, 1] = np.clip((0.25 * images["HH"] + 0.25 * images["HV"] - 0.25 * images["VH"] - 0.25 * images["VV"]), 0, None)  # M01
    M[:, :, 0, 2] = np.clip((0.25 * images["PH"] + 0.25 * images["PV"] - 0.25 * images["QH"] - 0.25 * images["QV"]), 0, None)  # M02
    M[:, :, 0, 3] = np.clip((0.25 * images["RH"] + 0.25 * images["RV"] - 0.25 * images["LH"] - 0.25 * images["LV"]), 0, None)  # M03

    M[:, :, 1, 0] = np.clip((0.25 * images["HH"] - 0.25 * images["HV"] + 0.25 * images["VH"] - 0.25 * images["VV"]), 0, None)  # M10
    M[:, :, 1, 1] = np.clip((0.25 * images["HH"] - 0.25 * images["HV"] - 0.25 * images["VH"] + 0.25 * images["VV"]), 0, None)  # M11
    M[:, :, 1, 2] = np.clip((0.25 * images["PH"] + 0.25 * images["QV"] - 0.25 * images["QH"] - 0.25 * images["PV"]), 0, None)  # M12
    M[:, :, 1, 3] = np.clip((0.25 * images["RH"] + 0.25 * images["LV"] - 0.25 * images["LH"] - 0.25 * images["RV"]), 0, None)  # M13

    M[:, :, 2, 0] = np.clip((0.25 * images["HP"] - 0.25 * images["HQ"] + 0.25 * images["VP"] - 0.25 * images["VQ"]), 0, None)  # M20
    M[:, :, 2, 1] = np.clip((0.25 * images["HP"] + 0.25 * images["VQ"] - 0.25 * images["VP"] - 0.25 * images["HQ"]), 0, None)  # M21
    M[:, :, 2, 2] = np.clip((0.25 * images["PP"] + 0.25 * images["QQ"] - 0.25 * images["QP"] - 0.25 * images["PQ"]), 0, None)  # M22
    M[:, :, 2, 3] = np.clip((0.25 * images["RP"] + 0.25 * images["LQ"] - 0.25 * images["LP"] - 0.25 * images["RQ"]), 0, None)  # M23

    M[:, :, 3, 0] = np.clip((0.25 * images["HR"] - 0.25 * images["HL"] + 0.25 * images["VR"] - 0.25 * images["VL"]), 0, None)  # M30
    M[:, :, 3, 1] = np.clip((0.25 * images["VL"] + 0.25 * images["HR"] - 0.25 * images["HL"] - 0.25 * images["VR"]), 0, None)  # M31
    M[:, :, 3, 2] = np.clip((0.25 * images["QL"] + 0.25 * images["PR"] - 0.25 * images["PL"] - 0.25 * images["QR"]), 0, None)  # M32
    M[:, :, 3, 3] = np.clip((0.25 * images["LL"] + 0.25 * images["RR"] - 0.25 * images["RL"] - 0.25 * images["LR"]), 0, None)  # M33

    # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –≤ –æ—Ç–¥–µ–ª—å–Ω—ã–µ —Ñ–∞–π–ª—ã

    os.makedirs(output_dir, exist_ok=True)

    for i in range(4):
        for j in range(4):
            output_image = Image.fromarray(M[:, :, i, j].astype(np.uint8))
            output_image.save(os.path.join(output_dir, f"M{i}{j}.png"))

    print(f"–í—Å–µ –æ–ø–µ—Ä–∞—Ü–∏–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω—ã –∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ –ø–∞–ø–∫—É {output_dir}.")
```

```python
import os
import numpy as np
from PIL import Image

# –ü—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω—ã–º–∏ –º–∞—Å—Å–∏–≤–∞–º–∏ M
input_dir = "HALF_RED_M"  # –ü–∞–ø–∫–∞, –≥–¥–µ –ª–µ–∂–∞—Ç M00, M01, M02, M03 –∏ —Ç.–¥.
output_dir = "HALF_RED_D"
# –ó–∞–≥—Ä—É–∑–∫–∞ –º–∞—Å—Å–∏–≤–æ–≤ M00, M01, M02, M03
M00 = np.array(Image.open(os.path.join(input_dir, "M00.png")))
M01 = np.array(Image.open(os.path.join(input_dir, "M01.png")))
M02 = np.array(Image.open(os.path.join(input_dir, "M02.png")))
M03 = np.array(Image.open(os.path.join(input_dir, "M03.png")))

# –ü—Ä–æ–≤–µ—Ä–∫–∞, —á—Ç–æ –º–∞—Å—Å–∏–≤—ã –∑–∞–≥—Ä—É–∂–µ–Ω—ã
if M00 is None or M01 is None or M02 is None or M03 is None:
    print("–û—à–∏–±–∫–∞: –Ω–µ —É–¥–∞–ª–æ—Å—å –∑–∞–≥—Ä—É–∑–∏—Ç—å –æ–¥–∏–Ω –∏–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ –º–∞—Å—Å–∏–≤–æ–≤.")
else:
    # –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ D
    # –ß—Ç–æ–±—ã –∏–∑–±–µ–∂–∞—Ç—å –¥–µ–ª–µ–Ω–∏—è –Ω–∞ –Ω–æ–ª—å, –¥–æ–±–∞–≤–∏–º –Ω–µ–±–æ–ª—å—à–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, 1e-10) –∫ M00
    epsilon = 1e-10
    D = (1 / (M00 + epsilon)) * np.sqrt(M01**2 + M02**2 + M03**2)

    # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞

    os.makedirs(output_dir, exist_ok=True)

    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º D –≤ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –∏ —Å–æ—Ö—Ä–∞–Ω—è–µ–º
    D_image = Image.fromarray((D * 255).astype(np.uint8))  # –ú–∞—Å—à—Ç–∞–±–∏—Ä—É–µ–º –∫ –¥–∏–∞–ø–∞–∑–æ–Ω—É 0-255
    D_image.save(os.path.join(output_dir, "D.png"))

    print(f"–ü–∞—Ä–∞–º–µ—Ç—Ä D —Ä–∞—Å—Å—á–∏—Ç–∞–Ω –∏ —Å–æ—Ö—Ä–∞–Ω—ë–Ω –≤ –ø–∞–ø–∫—É {output_dir}.")
```

```python
import os
import numpy as np
from PIL import Image

# –ü—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω—ã–º–∏ –º–∞—Å—Å–∏–≤–∞–º–∏ M
input_dir = "HALF_RED_M"  # –ü–∞–ø–∫–∞, –≥–¥–µ –ª–µ–∂–∞—Ç M00, M01, M02, M03 –∏ —Ç.–¥.
output_dir = "HALF_RED_P"
# –ó–∞–≥—Ä—É–∑–∫–∞ –º–∞—Å—Å–∏–≤–æ–≤ M00, M10, M20, M30
M00 = np.array(Image.open(os.path.join(input_dir, "M00.png")))
M10 = np.array(Image.open(os.path.join(input_dir, "M10.png")))
M20 = np.array(Image.open(os.path.join(input_dir, "M20.png")))
M30 = np.array(Image.open(os.path.join(input_dir, "M30.png")))

# –ü—Ä–æ–≤–µ—Ä–∫–∞, —á—Ç–æ –º–∞—Å—Å–∏–≤—ã –∑–∞–≥—Ä—É–∂–µ–Ω—ã
if M00 is None or M10 is None or M20 is None or M30 is None:
    print("–û—à–∏–±–∫–∞: –Ω–µ —É–¥–∞–ª–æ—Å—å –∑–∞–≥—Ä—É–∑–∏—Ç—å –æ–¥–∏–Ω –∏–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ –º–∞—Å—Å–∏–≤–æ–≤.")
else:
    # –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ D
    # –ß—Ç–æ–±—ã –∏–∑–±–µ–∂–∞—Ç—å –¥–µ–ª–µ–Ω–∏—è –Ω–∞ –Ω–æ–ª—å, –¥–æ–±–∞–≤–∏–º –Ω–µ–±–æ–ª—å—à–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, 1e-10) –∫ M00
    epsilon = 1e-10
    P = (1 / (M00 + epsilon)) * np.sqrt(M10**2 + M20**2 + M30**2)

    # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞

    os.makedirs(output_dir, exist_ok=True)

    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º P –≤ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –∏ —Å–æ—Ö—Ä–∞–Ω—è–µ–º
    P_image = Image.fromarray((P * 255).astype(np.uint8))  # –ú–∞—Å—à—Ç–∞–±–∏—Ä—É–µ–º –∫ –¥–∏–∞–ø–∞–∑–æ–Ω—É 0-255
    P_image.save(os.path.join(output_dir, "P.png"))

    print(f"–ü–∞—Ä–∞–º–µ—Ç—Ä P —Ä–∞—Å—Å—á–∏—Ç–∞–Ω –∏ —Å–æ—Ö—Ä–∞–Ω—ë–Ω –≤ –ø–∞–ø–∫—É {output_dir}.")
```

```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# –ü—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω—ã–º–∏ –º–∞—Å—Å–∏–≤–∞–º–∏ M
input_dir = "FRESH_RED_M"  # –ü–∞–ø–∫–∞, –≥–¥–µ –ª–µ–∂–∞—Ç M00, M01, M02, M03 –∏ —Ç.–¥.

# –ó–∞–≥—Ä—É–∑–∫–∞ –º–∞—Å—Å–∏–≤–æ–≤ M10, M20, M30, M01, M02, M03
M10 = np.array(Image.open(os.path.join(input_dir, "M10.png")))
M20 = np.array(Image.open(os.path.join(input_dir, "M20.png")))
M30 = np.array(Image.open(os.path.join(input_dir, "M30.png")))
M01 = np.array(Image.open(os.path.join(input_dir, "M01.png")))
M02 = np.array(Image.open(os.path.join(input_dir, "M02.png")))
M03 = np.array(Image.open(os.path.join(input_dir, "M03.png")))

# –ü—Ä–æ–≤–µ—Ä–∫–∞, —á—Ç–æ –º–∞—Å—Å–∏–≤—ã –∑–∞–≥—Ä—É–∂–µ–Ω—ã –∏ –Ω–µ –ø—É—Å—Ç—ã
if (M10.size == 0 or M20.size == 0 or M30.size == 0 or
    M01.size == 0 or M02.size == 0 or M03.size == 0):
    print("–û—à–∏–±–∫–∞: –æ–¥–∏–Ω –∏–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ –º–∞—Å—Å–∏–≤–æ–≤ –ø—É—Å—Ç—ã.")
else:
    # –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—É–º–º –∫–≤–∞–¥—Ä–∞—Ç–æ–≤
    sum_left = M10**2 + M20**2 + M30**2  # –õ–µ–≤–∞—è —á–∞—Å—Ç—å —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è
    sum_right = M01**2 + M02**2 + M03**2  # –ü—Ä–∞–≤–∞—è —á–∞—Å—Ç—å —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è

    # –†–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É —Å—É–º–º–∞–º–∏
    difference = sum_left - sum_right

    # –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏
    def normalize(data):
        return (data - data.min()) / (data.max() - data.min())

    sum_left_norm = normalize(sum_left)
    sum_right_norm = normalize(sum_right)
    difference_norm = normalize(difference)

    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å –ø–æ–º–æ—â—å—é matplotlib
    plt.figure(figsize=(15, 5))

    # –õ–µ–≤–∞—è —á–∞—Å—Ç—å
    plt.subplot(1, 3, 1)
    plt.imshow(sum_left_norm, cmap='viridis')
    plt.title("–õ–µ–≤–∞—è —á–∞—Å—Ç—å (M10^2 + M20^2 + M30^2)")
    plt.colorbar()

    # –ü—Ä–∞–≤–∞—è —á–∞—Å—Ç—å
    plt.subplot(1, 3, 2)
    plt.imshow(sum_right_norm, cmap='viridis')
    plt.title("–ü—Ä–∞–≤–∞—è —á–∞—Å—Ç—å (M01^2 + M02^2 + M03^2)")
    plt.colorbar()

    # –†–∞–∑–Ω–∏—Ü–∞
    plt.subplot(1, 3, 3)
    plt.imshow(difference_norm, cmap='coolwarm')
    plt.title("–†–∞–∑–Ω–∏—Ü–∞ (–õ–µ–≤–∞—è —á–∞—Å—Ç—å - –ü—Ä–∞–≤–∞—è —á–∞—Å—Ç—å)")
    plt.colorbar()

    # –ü–æ–∫–∞–∑–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏
    plt.tight_layout()
    plt.show()

    # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –≤ –≤–∏–¥–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
    output_dir = "FRESH_RED_COMPARISON"
    os.makedirs(output_dir, exist_ok=True)

    # –°–æ—Ö—Ä–∞–Ω—è–µ–º –ª–µ–≤—É—é —á–∞—Å—Ç—å
    sum_left_image = Image.fromarray((sum_left_norm * 255).astype(np.uint8))
    sum_left_image.save(os.path.join(output_dir, "sum_left.png"))

    # –°–æ—Ö—Ä–∞–Ω—è–µ–º –ø—Ä–∞–≤—É—é —á–∞—Å—Ç—å
    sum_right_image = Image.fromarray((sum_right_norm * 255).astype(np.uint8))
    sum_right_image.save(os.path.join(output_dir, "sum_right.png"))

    # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ä–∞–∑–Ω–∏—Ü—É
    difference_image = Image.fromarray((difference_norm * 255).astype(np.uint8))
    difference_image.save(os.path.join(output_dir, "difference.png"))

    print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ –ø–∞–ø–∫—É {output_dir}.")
```

```python
import os
import numpy as np
from PIL import Image

# –ü—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏ M00 ... M33
input_folder = "/content/HALF_RED_M"  # –ó–∞–º–µ–Ω–∏ –Ω–∞ –ø—É—Ç—å –∫ —Å–≤–æ–µ–π –ø–∞–ø–∫–µ
output_folder = "/content/HALF_RED_H(M)"  # –ü–∞–ø–∫–∞ –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è H00 ... H33

# –°–æ–∑–¥–∞–µ–º –ø–∞–ø–∫—É –¥–ª—è –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö, –µ—Å–ª–∏ –µ—ë –Ω–µ—Ç
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# –ó–∞–≥—Ä—É–∂–∞–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è M00 ... M33
M = {}
for i in range(4):
    for j in range(4):
        filename = f"M{i}{j}.png"
        filepath = os.path.join(input_folder, filename)
        if os.path.exists(filepath):
            img = Image.open(filepath)
            M[f"M{i}{j}"] = np.array(img) / 255.0  # –ù–æ—Ä–º–∞–ª–∏–∑—É–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –¥–æ [0, 1]
        else:
            raise FileNotFoundError(f"–§–∞–π–ª {filename} –Ω–µ –Ω–∞–π–¥–µ–Ω!")

# –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç—ã–µ –º–∞—Å—Å–∏–≤—ã –¥–ª—è H00 ... H33
H = {}
for i in range(4):
    for j in range(4):
        H[f"H{i}{j}"] = np.zeros_like(M["M00"])

# –í—ã—á–∏—Å–ª—è–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã H(M) –ø–æ —Ñ–æ—Ä–º—É–ª–∞–º
H["H00"] = 0.25 * (M["M00"] + M["M11"] + M["M22"] + M["M33"])
H["H01"] = 0.25 * (M["M01"] + M["M12"] + M["M23"] + M["M30"])
H["H02"] = 0.25 * (M["M02"] + M["M13"] - M["M20"] - M["M31"])
H["H03"] = 0.25 * (M["M03"] - M["M10"] + M["M21"] - M["M32"])
H["H10"] = H["H01"]  # –°–∏–º–º–µ—Ç—Ä–∏—è
H["H11"] = 0.25 * (M["M00"] + M["M11"] - M["M22"] - M["M33"])
H["H12"] = 0.25 * (M["M03"] + M["M10"] + M["M21"] + M["M32"])
H["H13"] = 0.25 * (M["M02"] - M["M13"] + M["M20"] - M["M31"])
H["H20"] = H["H02"]  # –°–∏–º–º–µ—Ç—Ä–∏—è
H["H21"] = H["H12"]  # –°–∏–º–º–µ—Ç—Ä–∏—è
H["H22"] = 0.25 * (M["M00"] - M["M11"] + M["M22"] - M["M33"])
H["H23"] = 0.25 * (M["M01"] - M["M12"] - M["M23"] + M["M30"])
H["H30"] = H["H03"]  # –°–∏–º–º–µ—Ç—Ä–∏—è
H["H31"] = H["H13"]  # –°–∏–º–º–µ—Ç—Ä–∏—è
H["H32"] = H["H23"]  # –°–∏–º–º–µ—Ç—Ä–∏—è
H["H33"] = 0.25 * (M["M00"] - M["M11"] - M["M22"] + M["M33"])

# –ü—Ä–∏–º–µ–Ω—è–µ–º np.clip –¥–ª—è –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è –∑–Ω–∞—á–µ–Ω–∏–π
for key in H:
    H[key] = np.clip(H[key], 0, 1)  # –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –¥–æ [0, 1]

# –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤ –Ω–æ–≤—É—é –ø–∞–ø–∫—É
for i in range(4):
    for j in range(4):
        filename = f"H{i}{j}.png"
        filepath = os.path.join(output_folder, filename)
        img = Image.fromarray((H[f"H{i}{j}"] * 255).astype(np.uint8))  # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –æ–±—Ä–∞—Ç–Ω–æ –≤ [0, 255]
        img.save(filepath)

print("–í—Å–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è H00 ... H33 —É—Å–ø–µ—à–Ω–æ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã!")
```

```python
import os
import numpy as np
from PIL import Image

# –ü—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏ H00 ... H33
input_folder = "/content/HALF_RED_H(M)"  # –ó–∞–º–µ–Ω–∏ –Ω–∞ –ø—É—Ç—å –∫ —Å–≤–æ–µ–π –ø–∞–ø–∫–µ
output_folder = "/content/HALF_RED_P1P2P3"  # –ü–∞–ø–∫–∞ –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è P1, P2, P3

# –°–æ–∑–¥–∞–µ–º –ø–∞–ø–∫—É –¥–ª—è –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö, –µ—Å–ª–∏ –µ—ë –Ω–µ—Ç
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# –ó–∞–≥—Ä—É–∂–∞–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è H00 ... H33
H = {}
for i in range(4):
    for j in range(4):
        filename = f"H{i}{j}.png"
        filepath = os.path.join(input_folder, filename)
        if os.path.exists(filepath):
            img = Image.open(filepath)
            H[f"H{i}{j}"] = np.array(img) / 255.0  # –ù–æ—Ä–º–∞–ª–∏–∑—É–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –¥–æ [0, 1]
        else:
            raise FileNotFoundError(f"–§–∞–π–ª {filename} –Ω–µ –Ω–∞–π–¥–µ–Ω!")

# –í—ã—á–∏—Å–ª—è–µ–º —Å–ª–µ–¥ (—Å—É–º–º—É –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤)
trace = H["H00"] + H["H11"] + H["H22"] + H["H33"]

# –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç—ã–µ –º–∞—Å—Å–∏–≤—ã –¥–ª—è P1, P2, P3
P1 = np.zeros_like(H["H00"])
P2 = np.zeros_like(H["H00"])
P3 = np.zeros_like(H["H00"])

# –†–∞–∑–º–µ—Ä—ã –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
height, width = H["H00"].shape

# –í—ã—á–∏—Å–ª—è–µ–º P1, P2, P3 –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–∏–∫—Å–µ–ª—è
for i in range(height):
    for j in range(width):
        # –°–æ–∑–¥–∞–µ–º –º–∞—Ç—Ä–∏—Ü—É H(M) –¥–ª—è —Ç–µ–∫—É—â–µ–≥–æ –ø–∏–∫—Å–µ–ª—è
        H_pixel = np.array([
            [H["H00"][i, j], H["H01"][i, j], H["H02"][i, j], H["H03"][i, j]],
            [H["H10"][i, j], H["H11"][i, j], H["H12"][i, j], H["H13"][i, j]],
            [H["H20"][i, j], H["H21"][i, j], H["H22"][i, j], H["H23"][i, j]],
            [H["H30"][i, j], H["H31"][i, j], H["H32"][i, j], H["H33"][i, j]]
        ])

        # –í—ã—á–∏—Å–ª—è–µ–º —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
        eigenvalues = np.linalg.eigvalsh(H_pixel)  # –ò—Å–ø–æ–ª—å–∑—É–µ–º eigvalsh –¥–ª—è —ç—Ä–º–∏—Ç–æ–≤—ã—Ö –º–∞—Ç—Ä–∏—Ü
        eigenvalues = np.sort(eigenvalues)[::-1]  # –°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é

        # –í—ã—á–∏—Å–ª—è–µ–º P1, P2, P3
        if trace[i, j] != 0:  # –ò–∑–±–µ–≥–∞–µ–º –¥–µ–ª–µ–Ω–∏—è –Ω–∞ –Ω–æ–ª—å
            P1[i, j] = (eigenvalues[0] - eigenvalues[1]) / trace[i, j]
            P2[i, j] = (eigenvalues[0] + eigenvalues[1] - 2 * eigenvalues[2]) / trace[i, j]
            P3[i, j] = (eigenvalues[0] + eigenvalues[1] + eigenvalues[2] - 3 * eigenvalues[3]) / trace[i, j]

# –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è P1, P2, P3 –¥–∏–∞–ø–∞–∑–æ–Ω–æ–º [0, 1] —Å –ø–æ–º–æ—â—å—é np.clip
P1 = np.clip(P1, 0, 1)
P2 = np.clip(P2, 0, 1)
P3 = np.clip(P3, 0, 1)

# –°–æ—Ö—Ä–∞–Ω—è–µ–º P1, P2, P3 –≤ –≤–∏–¥–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
P1_img = Image.fromarray((P1 * 255).astype(np.uint8))
P1_img.save(os.path.join(output_folder, "P1.png"))

P2_img = Image.fromarray((P2 * 255).astype(np.uint8))
P2_img.save(os.path.join(output_folder, "P2.png"))

P3_img = Image.fromarray((P3 * 255).astype(np.uint8))
P3_img.save(os.path.join(output_folder, "P3.png"))

print("–í—Å–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è P1, P2, P3 —É—Å–ø–µ—à–Ω–æ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã!")
```

```python
import os
import numpy as np
from PIL import Image

# –ü—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏ M00 ... M33
input_folder = "/content/HALF_RED_M"  # –ó–∞–º–µ–Ω–∏ –Ω–∞ –ø—É—Ç—å –∫ —Å–≤–æ–µ–π –ø–∞–ø–∫–µ
output_folder = "/content/HALF_RED_Ps"  # –ü–∞–ø–∫–∞ –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è Ps

# –°–æ–∑–¥–∞–µ–º –ø–∞–ø–∫—É –¥–ª—è –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö, –µ—Å–ª–∏ –µ—ë –Ω–µ—Ç
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# –ó–∞–≥—Ä—É–∂–∞–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è M00 ... M33
M = {}
for i in range(4):
    for j in range(4):
        filename = f"M{i}{j}.png"
        filepath = os.path.join(input_folder, filename)
        if os.path.exists(filepath):
            img = Image.open(filepath)
            M[f"M{i}{j}"] = np.array(img) / 255.0  # –ù–æ—Ä–º–∞–ª–∏–∑—É–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –¥–æ [0, 1]
        else:
            raise FileNotFoundError(f"–§–∞–π–ª {filename} –Ω–µ –Ω–∞–π–¥–µ–Ω!")

# –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π –º–∞—Å—Å–∏–≤ –¥–ª—è Ps
Ps = np.zeros_like(M["M00"])

# –†–∞–∑–º–µ—Ä—ã –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
height, width = M["M00"].shape

# –í—ã—á–∏—Å–ª—è–µ–º Ps –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–∏–∫—Å–µ–ª—è
for i in range(height):
    for j in range(width):
        # –°–æ–∑–¥–∞–µ–º –ø–æ–¥–º–∞—Ç—Ä–∏—Ü—É m
        m = np.array([
            [M["M11"][i, j], M["M12"][i, j], M["M13"][i, j]],
            [M["M21"][i, j], M["M22"][i, j], M["M23"][i, j]],
            [M["M31"][i, j], M["M32"][i, j], M["M33"][i, j]]
        ])

        # –í—ã—á–∏—Å–ª—è–µ–º 2-–Ω–æ—Ä–º—É (—Å–ø–µ–∫—Ç—Ä–∞–ª—å–Ω—É—é –Ω–æ—Ä–º—É) –ø–æ–¥–º–∞—Ç—Ä–∏—Ü—ã m
        norm_m = np.linalg.norm(m, ord=2)

        # –í—ã—á–∏—Å–ª—è–µ–º Ps
        Ps[i, j] = norm_m / np.sqrt(3)

# –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è Ps –¥–∏–∞–ø–∞–∑–æ–Ω–æ–º [0, 1] —Å –ø–æ–º–æ—â—å—é np.clip
Ps = np.clip(Ps, 0, 1)

# –°–æ—Ö—Ä–∞–Ω—è–µ–º Ps –≤ –≤–∏–¥–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
Ps_img = Image.fromarray((Ps * 255).astype(np.uint8))
Ps_img.save(os.path.join(output_folder, "Ps.png"))

print("–ò–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ Ps —É—Å–ø–µ—à–Ω–æ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–æ!")
```

```python
import os
import numpy as np
from PIL import Image

# –ü—É—Ç–∏ –∫ –ø–∞–ø–∫–∞–º —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏
input_folder_D = "/content/HALF_RED_D"  # –ü–∞–ø–∫–∞ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ–º D
input_folder_P = "/content/HALF_RED_P"  # –ü–∞–ø–∫–∞ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ–º P
input_folder_Ps = "/content/HALF_RED_Ps"  # –ü–∞–ø–∫–∞ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ–º Ps
input_folder_P1P2P3 = "/content/HALF_RED_P1P2P3"  # –ü–∞–ø–∫–∞ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏ P1, P2, P3
output_folder = "/content/HALF_RED_PDelta"  # –ü–∞–ø–∫–∞ –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è PŒî

# –°–æ–∑–¥–∞–µ–º –ø–∞–ø–∫—É –¥–ª—è –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö, –µ—Å–ª–∏ –µ—ë –Ω–µ—Ç
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# –ó–∞–≥—Ä—É–∂–∞–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
def load_image(folder, filename):
    filepath = os.path.join(folder, filename)
    if os.path.exists(filepath):
        img = Image.open(filepath)
        return np.array(img) / 255.0  # –ù–æ—Ä–º–∞–ª–∏–∑—É–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –¥–æ [0, 1]
    else:
        raise FileNotFoundError(f"–§–∞–π–ª {filename} –Ω–µ –Ω–∞–π–¥–µ–Ω!")

D = load_image(input_folder_D, "D.png")
P = load_image(input_folder_P, "P.png")
Ps = load_image(input_folder_Ps, "Ps.png")
P1 = load_image(input_folder_P1P2P3, "P1.png")
P2 = load_image(input_folder_P1P2P3, "P2.png")
P3 = load_image(input_folder_P1P2P3, "P3.png")

# –í—ã—á–∏—Å–ª—è–µ–º PŒî –ø–µ—Ä–≤—ã–º —Å–ø–æ—Å–æ–±–æ–º
P_delta_method1 = (1 / np.sqrt(3)) * np.sqrt(2 * P1**2 + (2/3) * P2**2 + (1/3) * P3**2)

# –í—ã—á–∏—Å–ª—è–µ–º PŒî –≤—Ç–æ—Ä—ã–º —Å–ø–æ—Å–æ–±–æ–º
P_delta_method2 = (1 / 3) * np.sqrt(D**2 + P**2 + 3 * Ps**2)

# –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è PŒî –¥–∏–∞–ø–∞–∑–æ–Ω–æ–º [0, 1] —Å –ø–æ–º–æ—â—å—é np.clip
P_delta_method1 = np.clip(P_delta_method1, 0, 1)
P_delta_method2 = np.clip(P_delta_method2, 0, 1)

# –°–æ—Ö—Ä–∞–Ω—è–µ–º PŒî –≤ –≤–∏–¥–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
P_delta_method1_img = Image.fromarray((P_delta_method1 * 255).astype(np.uint8))
P_delta_method1_img.save(os.path.join(output_folder, "P_delta_method1.png"))

P_delta_method2_img = Image.fromarray((P_delta_method2 * 255).astype(np.uint8))
P_delta_method2_img.save(os.path.join(output_folder, "P_delta_method2.png"))

print("–ò–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è PŒî —É—Å–ø–µ—à–Ω–æ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã!")
```

```python
import os
import numpy as np
from scipy.linalg import polar
from PIL import Image

# –ü—É—Ç–∏ –∫ –ø–∞–ø–∫–∞–º
input_folder = "/content/HALF_RED_M"  # –ó–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –ø—É—Ç—å –∫ –≤–∞—à–µ–π –ø–∞–ø–∫–µ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏
output_folder = "/content/HALF_RED_MR"      # –ó–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –ø—É—Ç—å –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏–∏ –º–∞—Ç—Ä–∏—Ü—ã –ú—é–ª–ª–µ—Ä–∞ –∏ –≤—ã–¥–µ–ª–µ–Ω–∏—è M_R
def extract_MR(M):
    """
    –í—ã–¥–µ–ª—è–µ—Ç –º–∞—Ç—Ä–∏—Ü—É M_R –∏–∑ –º–∞—Ç—Ä–∏—Ü—ã –ú—é–ª–ª–µ—Ä–∞ M.
    """
    U, _ = polar(M[1:4, 1:4])  # –ü–æ–ª—è—Ä–Ω–æ–µ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ –¥–ª—è –ø–æ–¥–º–∞—Ç—Ä–∏—Ü—ã 3x3
    M_R = np.eye(4)
    M_R[1:4, 1:4] = U  # –ó–∞–º–µ–Ω—è–µ–º –ø–æ–¥–º–∞—Ç—Ä–∏—Ü—É –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Ä–∞–∑–ª–æ–∂–µ–Ω–∏—è
    return M_R

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
def process_mueller_images(input_folder, output_folder):
    """
    –ß–∏—Ç–∞–µ—Ç –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –∏–∑ input_folder, –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –∏—Ö –∏ —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç M_R –≤ output_folder.
    """
    # –°–æ–∑–¥–∞–µ–º –≤—ã—Ö–æ–¥–Ω—É—é –ø–∞–ø–∫—É, –µ—Å–ª–∏ –µ—ë –Ω–µ—Ç
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # –°–ø–∏—Å–æ–∫ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Ç—Ä–∏—Ü—ã –ú—é–ª–ª–µ—Ä–∞
    elements = ["M00", "M01", "M02", "M03",
               "M10", "M11", "M12", "M13",
               "M20", "M21", "M22", "M23",
               "M30", "M31", "M32", "M33"]

    # –ó–∞–≥—Ä—É–∂–∞–µ–º –≤—Å–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –∏ –æ–±—ä–µ–¥–∏–Ω—è–µ–º –∏—Ö –≤ –æ–¥–Ω—É –º–∞—Ç—Ä–∏—Ü—É –ú—é–ª–ª–µ—Ä–∞
    mueller_images = {}
    for element in elements:
        file_path = os.path.join(input_folder, f"{element}.png")
        if os.path.exists(file_path):
            image = Image.open(file_path)
            mueller_images[element] = np.array(image)
        else:
            raise FileNotFoundError(f"–§–∞–π–ª {file_path} –Ω–µ –Ω–∞–π–¥–µ–Ω.")

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–∞–∑–º–µ—Ä—ã –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
    height, width = mueller_images["M00"].shape
    for element in elements:
        if mueller_images[element].shape != (height, width):
            raise ValueError(f"–ò–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ {element} –∏–º–µ–µ—Ç –Ω–µ–≤–µ—Ä–Ω—ã–π —Ä–∞–∑–º–µ—Ä.")

    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –º–∞—Å—Å–∏–≤–∞ –¥–ª—è –º–∞—Ç—Ä–∏—Ü—ã –ú—é–ª–ª–µ—Ä–∞ (250, 250, 4, 4)
    M = np.zeros((height, width, 4, 4), dtype=np.float32)

    # –ó–∞–ø–æ–ª–Ω—è–µ–º –º–∞—Ç—Ä–∏—Ü—É –ú—é–ª–ª–µ—Ä–∞
    for i, element in enumerate(elements):
        row = i // 4
        col = i % 4
        M[:, :, row, col] = mueller_images[element]

    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –º–∞—Å—Å–∏–≤–∞ –¥–ª—è M_R (250, 250, 4, 4)
    M_R = np.zeros((height, width, 4, 4), dtype=np.float32)

    # –û–±—Ä–∞–±–æ—Ç–∫–∞ –∫–∞–∂–¥–æ–≥–æ –ø–∏–∫—Å–µ–ª—è
    for i in range(height):
        for j in range(width):
            # –ò–∑–≤–ª–µ–∫–∞–µ–º –º–∞—Ç—Ä–∏—Ü—É –ú—é–ª–ª–µ—Ä–∞ –¥–ª—è –ø–∏–∫—Å–µ–ª—è (i, j)
            M_pixel = M[i, j]

            # –í—ã–¥–µ–ª—è–µ–º M_R
            M_R_pixel = extract_MR(M_pixel)

            # –°–æ—Ö—Ä–∞–Ω—è–µ–º M_R –≤ –º–∞—Å—Å–∏–≤
            M_R[i, j] = M_R_pixel

    # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ (–∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç M_R –≤ –æ—Ç–¥–µ–ª—å–Ω—ã–π —Ñ–∞–π–ª)
    for i, element in enumerate(elements):
        row = i // 4
        col = i % 4
        MR_element = M_R[:, :, row, col]

        # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
        if np.all(MR_element == MR_element[0, 0]):  # –í—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è –æ–¥–∏–Ω–∞–∫–æ–≤—ã
            MR_element_normalized = np.zeros_like(MR_element, dtype=np.uint8)  # –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –≤ 0
        else:
            # –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö –∏ –æ–±—Ä–µ–∑–∫–∞ –∑–Ω–∞—á–µ–Ω–∏–π
            MR_element_normalized = (MR_element - MR_element.min()) / (MR_element.max() - MR_element.min()) * 255
            MR_element_normalized = np.clip(MR_element_normalized, 0, 255)  # –û–±—Ä–µ–∑–∫–∞ –∑–Ω–∞—á–µ–Ω–∏–π
            MR_element_normalized = MR_element_normalized.astype(np.uint8)

        # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ
        output_path = os.path.join(output_folder, f"MR_{element}.png")
        Image.fromarray(MR_element_normalized).save(output_path)
        print(f"–°–æ—Ö—Ä–∞–Ω–µ–Ω–æ: {output_path}")

# –ó–∞–ø—É—Å–∫ –æ–±—Ä–∞–±–æ—Ç–∫–∏
process_mueller_images(input_folder, output_folder)
```

```python
import os
import numpy as np
from PIL import Image

# –£–∫–∞–∂–∏—Ç–µ –ø—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏ MR
input_folder = "/content/HALF_RED_MR"  # –ó–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –ø—É—Ç—å –∫ –≤–∞—à–µ–π –ø–∞–ø–∫–µ —Å MR_M11.png, MR_M22.png, MR_M33.png
output_folder = "/content/HALF_RED_R"  # –£–∫–∞–∂–∏—Ç–µ –ø—É—Ç—å –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
output_path = os.path.join(output_folder, "R_output.png")  # –ü–æ–ª–Ω—ã–π –ø—É—Ç—å –∫ –≤—ã—Ö–æ–¥–Ω–æ–º—É —Ñ–∞–π–ª—É

# –°–æ–∑–¥–∞–π—Ç–µ –ø–∞–ø–∫—É –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è, –µ—Å–ª–∏ –µ—ë –Ω–µ—Ç
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# –ó–∞–≥—Ä—É–∑–∏—Ç–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è MR_M11, MR_M22, MR_M33
mr_m11_path = os.path.join(input_folder, "MR_M11.png")
mr_m22_path = os.path.join(input_folder, "MR_M22.png")
mr_m33_path = os.path.join(input_folder, "MR_M33.png")

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è —Ñ–∞–π–ª–æ–≤
if not os.path.exists(mr_m11_path) or not os.path.exists(mr_m22_path) or not os.path.exists(mr_m33_path):
    raise FileNotFoundError("–û–¥–∏–Ω –∏–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ñ–∞–π–ª–æ–≤ MR_M11.png, MR_M22.png, MR_M33.png –Ω–µ –Ω–∞–π–¥–µ–Ω—ã –≤ —É–∫–∞–∑–∞–Ω–Ω–æ–π –ø–∞–ø–∫–µ.")

# –ó–∞–≥—Ä—É–∑–∏—Ç–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –∏ –Ω–æ—Ä–º–∞–ª–∏–∑—É–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è –∫ –¥–∏–∞–ø–∞–∑–æ–Ω—É [0, 1]
mr_m11 = np.array(Image.open(mr_m11_path), dtype=np.float32) / 255.0
mr_m22 = np.array(Image.open(mr_m22_path), dtype=np.float32) / 255.0
mr_m33 = np.array(Image.open(mr_m33_path), dtype=np.float32) / 255.0

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–∞–∑–º–µ—Ä–æ–≤ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
if mr_m11.shape != mr_m22.shape or mr_m11.shape != mr_m33.shape:
    raise ValueError("–ò–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –¥–æ–ª–∂–Ω—ã –∏–º–µ—Ç—å –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ —Ä–∞–∑–º–µ—Ä—ã.")

# –í—ã—á–∏—Å–ª–∏—Ç–µ R
trace = 1 + mr_m11 + mr_m22 + mr_m33  # –°–ª–µ–¥ –º–∞—Ç—Ä–∏—Ü—ã M_R
arg = np.abs((trace / 2) - 1)  # –ê—Ä–≥—É–º–µ–Ω—Ç –¥–ª—è arccos

# –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ [0, 1]. –í–∞–∂–Ω–æ –¥–µ–ª–∞—Ç—å —ç—Ç–æ –î–û arccos
arg = np.clip(arg, -1, 1)

R = np.arccos(arg)  # –§–æ—Ä–º—É–ª–∞ –¥–ª—è R

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ NaN –∏ inf
if np.any(np.isnan(R)) or np.any(np.isinf(R)):
    print("–ü—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–µ: –í –º–∞—Å—Å–∏–≤–µ R –æ–±–Ω–∞—Ä—É–∂–µ–Ω—ã NaN –∏–ª–∏ inf. –ó–∞–º–µ–Ω—è–µ–º –∏—Ö –Ω–∞ 0.")
    R = np.nan_to_num(R, nan=0.0, posinf=0.0, neginf=0.0)  # –ó–∞–º–µ–Ω—è–µ–º NaN –∏ inf –Ω–∞ 0

# –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
if np.all(R == R[0, 0]):  # –ï—Å–ª–∏ –≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è –æ–¥–∏–Ω–∞–∫–æ–≤—ã
    print("–ü—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–µ: –í—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è R –æ–¥–∏–Ω–∞–∫–æ–≤—ã. –†–µ–∑—É–ª—å—Ç–∞—Ç –±—É–¥–µ—Ç —á—ë—Ä–Ω—ã–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ–º.")
    R_normalized = np.zeros_like(R, dtype=np.uint8)
else:
    R_normalized = (R - R.min()) / (R.max() - R.min()) * 255  # –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –≤ –¥–∏–∞–ø–∞–∑–æ–Ω [0, 255]
    R_normalized = np.clip(R_normalized, 0, 255).astype(np.uint8)  # –û–±—Ä–µ–∑–∫–∞ –∑–Ω–∞—á–µ–Ω–∏–π

# –°–æ—Ö—Ä–∞–Ω–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
Image.fromarray(R_normalized).save(output_path)
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç —Å–æ—Ö—Ä–∞–Ω—ë–Ω –≤ {output_path}")
```

```python
import os
import numpy as np
from PIL import Image

# –£–∫–∞–∂–∏—Ç–µ –ø—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏ MR
input_folder = "/content/HALF_RED_MR"  # –ó–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –ø—É—Ç—å –∫ –≤–∞—à–µ–π –ø–∞–ø–∫–µ —Å MR_M11.png, MR_M22.png, –∏ —Ç.–¥.
output_folder = "/content/HALF_RED_Delta"  # –£–∫–∞–∂–∏—Ç–µ –ø—É—Ç—å –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
output_path = os.path.join(output_folder, "Delta_output.png")  # –ü–æ–ª–Ω—ã–π –ø—É—Ç—å –∫ –≤—ã—Ö–æ–¥–Ω–æ–º—É —Ñ–∞–π–ª—É

# –°–æ–∑–¥–∞–π—Ç–µ –ø–∞–ø–∫—É –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è, –µ—Å–ª–∏ –µ—ë –Ω–µ—Ç
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# –ó–∞–≥—Ä—É–∑–∏—Ç–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è MR11, MR22, MR12, MR21
mr_11_path = os.path.join(input_folder, "MR_M11.png")
mr_22_path = os.path.join(input_folder, "MR_M22.png")
mr_12_path = os.path.join(input_folder, "MR_M12.png") # –î–æ–±–∞–≤–ª–µ–Ω–æ MR12
mr_21_path = os.path.join(input_folder, "MR_M21.png") # –î–æ–±–∞–≤–ª–µ–Ω–æ MR21


# –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è —Ñ–∞–π–ª–æ–≤
if not os.path.exists(mr_11_path) or not os.path.exists(mr_22_path) or not os.path.exists(mr_12_path) or not os.path.exists(mr_21_path):
    raise FileNotFoundError("–û–¥–∏–Ω –∏–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ñ–∞–π–ª–æ–≤ MR_M11.png, MR_M22.png, MR_M12.png, MR_M21.png –Ω–µ –Ω–∞–π–¥–µ–Ω—ã –≤ —É–∫–∞–∑–∞–Ω–Ω–æ–π –ø–∞–ø–∫–µ.")

# –ó–∞–≥—Ä—É–∑–∏—Ç–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –∏ –Ω–æ—Ä–º–∞–ª–∏–∑—É–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è –∫ –¥–∏–∞–ø–∞–∑–æ–Ω—É [0, 1]
mr_11 = np.array(Image.open(mr_11_path), dtype=np.float32) / 255.0
mr_22 = np.array(Image.open(mr_22_path), dtype=np.float32) / 255.0
mr_12 = np.array(Image.open(mr_12_path), dtype=np.float32) / 255.0
mr_21 = np.array(Image.open(mr_21_path), dtype=np.float32) / 255.0


# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–∞–∑–º–µ—Ä–æ–≤ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
if not (mr_11.shape == mr_22.shape == mr_12.shape == mr_21.shape):
    raise ValueError("–ò–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –¥–æ–ª–∂–Ω—ã –∏–º–µ—Ç—å –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ —Ä–∞–∑–º–µ—Ä—ã.")


# –í—ã—á–∏—Å–ª–∏—Ç–µ Œ¥
term1 = (mr_11 + mr_22)**2
term2 = (mr_21 + mr_12)**2
argument = np.sqrt(term1 + term2) - 1

#–û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –∞—Ä–≥—É–º–µ–Ω—Ç —Ñ—É–Ω–∫—Ü–∏–∏ arccos. –í–∞–∂–Ω–æ! –î–∏–∞–ø–∞–∑–æ–Ω –æ—Ç -1 –¥–æ 1.
argument = np.clip(argument, -1, 1)

delta = np.arccos(argument) # –í—ã—á–∏—Å–ª–µ–Ω–∏–µ delta

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ NaN –∏ inf
if np.any(np.isnan(delta)) or np.any(np.isinf(delta)):
    print("–ü—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–µ: –í –º–∞—Å—Å–∏–≤–µ delta –æ–±–Ω–∞—Ä—É–∂–µ–Ω—ã NaN –∏–ª–∏ inf. –ó–∞–º–µ–Ω—è–µ–º –∏—Ö –Ω–∞ 0.")
    delta = np.nan_to_num(delta, nan=0.0, posinf=0.0, neginf=0.0)

# –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
if np.all(delta == delta[0, 0]):
    print("–ü—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–µ: –í—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è delta –æ–¥–∏–Ω–∞–∫–æ–≤—ã. –†–µ–∑—É–ª—å—Ç–∞—Ç –±—É–¥–µ—Ç —á—ë—Ä–Ω—ã–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ–º.")
    delta_normalized = np.zeros_like(delta, dtype=np.uint8)
else:
    delta_normalized = (delta - delta.min()) / (delta.max() - delta.min()) * 255
    delta_normalized = np.clip(delta_normalized, 0, 255).astype(np.uint8)

# –°–æ—Ö—Ä–∞–Ω–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
Image.fromarray(delta_normalized).save(output_path)
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç —Å–æ—Ö—Ä–∞–Ω—ë–Ω –≤ {output_path}")
```

```python
import os
import numpy as np
from PIL import Image

# –£–∫–∞–∂–∏—Ç–µ –ø—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏ MR
input_folder = "/content/HALF_RED_MR"  # –ó–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –ø—É—Ç—å –∫ –≤–∞—à–µ–π –ø–∞–ø–∫–µ —Å MR_M11.png, MR_M22.png, –∏ —Ç.–¥.
output_folder = "/content/HALF_RED_Psi"  # –£–∫–∞–∂–∏—Ç–µ –ø—É—Ç—å –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
output_path = os.path.join(output_folder, "Psi_output.png")  # –ü–æ–ª–Ω—ã–π –ø—É—Ç—å –∫ –≤—ã—Ö–æ–¥–Ω–æ–º—É —Ñ–∞–π–ª—É

# –°–æ–∑–¥–∞–π—Ç–µ –ø–∞–ø–∫—É –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è, –µ—Å–ª–∏ –µ—ë –Ω–µ—Ç
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# –ó–∞–≥—Ä—É–∑–∏—Ç–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è MR11, MR22, MR12, MR21
mr_11_path = os.path.join(input_folder, "MR_M11.png")
mr_22_path = os.path.join(input_folder, "MR_M22.png")
mr_12_path = os.path.join(input_folder, "MR_M12.png")
mr_21_path = os.path.join(input_folder, "MR_M21.png")

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è —Ñ–∞–π–ª–æ–≤
if not os.path.exists(mr_11_path) or not os.path.exists(mr_22_path) or not os.path.exists(mr_12_path) or not os.path.exists(mr_21_path):
    raise FileNotFoundError("–û–¥–∏–Ω –∏–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ñ–∞–π–ª–æ–≤ MR_M11.png, MR_M22.png, MR_M12.png, MR_M21.png –Ω–µ –Ω–∞–π–¥–µ–Ω—ã –≤ —É–∫–∞–∑–∞–Ω–Ω–æ–π –ø–∞–ø–∫–µ.")

# –ó–∞–≥—Ä—É–∑–∏—Ç–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –∏ –Ω–æ—Ä–º–∞–ª–∏–∑—É–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è –∫ –¥–∏–∞–ø–∞–∑–æ–Ω—É [0, 1]
mr_11 = np.array(Image.open(mr_11_path), dtype=np.float32) / 255.0
mr_22 = np.array(Image.open(mr_22_path), dtype=np.float32) / 255.0
mr_12 = np.array(Image.open(mr_12_path), dtype=np.float32) / 255.0
mr_21 = np.array(Image.open(mr_21_path), dtype=np.float32) / 255.0

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–∞–∑–º–µ—Ä–æ–≤ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
if not (mr_11.shape == mr_22.shape == mr_12.shape == mr_21.shape):
    raise ValueError("–ò–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –¥–æ–ª–∂–Ω—ã –∏–º–µ—Ç—å –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ —Ä–∞–∑–º–µ—Ä—ã.")

# –í—ã—á–∏—Å–ª–∏—Ç–µ œà
numerator = mr_21 - mr_12
denominator = mr_11 + mr_22

# –û–±—Ä–∞–±–æ—Ç–∫–∞ –¥–µ–ª–µ–Ω–∏—è –Ω–∞ –Ω–æ–ª—å.  –ï—Å–ª–∏ –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—å —Ä–∞–≤–µ–Ω 0, —É—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º psi = 0
psi = np.arctan2(numerator, denominator)

# `arctan2` –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ [-pi, pi].  –ù–æ—Ä–º–∞–ª–∏–∑—É–µ–º –∫ [0, 1].
psi_normalized_0_1 = (psi + np.pi) / (2 * np.pi)

# –ú–∞—Å—à—Ç–∞–±–∏—Ä—É–µ–º –∫ [0, 255] –∏ –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –∫ uint8
psi_normalized = (psi_normalized_0_1 * 255).astype(np.uint8)

# –°–æ—Ö—Ä–∞–Ω–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
Image.fromarray(psi_normalized).save(output_path)
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç —Å–æ—Ö—Ä–∞–Ω—ë–Ω –≤ {output_path}")
```

```python
from PIL import Image
import numpy as np
import os

def average_images(image1_path, image2_path, output_path):
    # –ó–∞–≥—Ä—É–∂–∞–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
    img1 = Image.open(image1_path).convert("L")  # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ –æ—Ç—Ç–µ–Ω–∫–∏ —Å–µ—Ä–æ–≥–æ
    img2 = Image.open(image2_path).convert("L")

    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ numpy –º–∞—Å—Å–∏–≤—ã
    arr1 = np.array(img1, dtype=np.float32)
    arr2 = np.array(img2, dtype=np.float32)

    # –°—Ä–µ–¥–Ω–µ–µ –¥–≤—É—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
    avg_arr = 0.5 * arr1 + 0.5 * arr2
    avg_arr = np.clip(avg_arr, 0, 255).astype(np.uint8)  # –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –∏ –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ uint8

    # –°–æ–∑–¥–∞—ë–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –∏ —Å–æ—Ö—Ä–∞–Ω—è–µ–º
    avg_img = Image.fromarray(avg_arr)
    avg_img.save(output_path)
    print(f"–°–æ—Ö—Ä–∞–Ω–µ–Ω–æ: {output_path}")

# –ü—É—Ç–∏ –∫ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º
image1_path = "/content/FRESH_GREEN_cropped/33.HL.png"  # –ó–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –Ω—É–∂–Ω—ã–π –ø—É—Ç—å
image2_path = "/content/SPOILT_GREEN_cropped/4.HL.png"  # –ó–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –Ω—É–∂–Ω—ã–π –ø—É—Ç—å
output_path = "33.HL.png"

# –í—ã–ø–æ–ª–Ω—è–µ–º —É—Å—Ä–µ–¥–Ω–µ–Ω–∏–µ
average_images(image1_path, image2_path, output_path)
```

```python
import os
import numpy as np
from PIL import Image
import glob
import re

# –ü—É—Ç–∏ –∫ –ø–∞–ø–∫–∞–º
fresh_folder = "/content/FRESH_GREEN_cropped"
spoiled_folder = "/content/SPOILT_GREEN_cropped"
half_fresh_folder = "/content/HALF_GREEN_cropped"

# –£–±–µ–¥–∏–º—Å—è, —á—Ç–æ –≤—ã—Ö–æ–¥–Ω–∞—è –ø–∞–ø–∫–∞ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç
os.makedirs(half_fresh_folder, exist_ok=True)

# –°–ø–∏—Å–æ–∫ –Ω–µ–¥–æ—Å—Ç–∞—é—â–∏—Ö –∏–∑–º–µ—Ä–µ–Ω–∏–π (–±–µ–∑ –Ω–æ–º–µ—Ä–æ–≤)
missing_patterns = ["LP", "LV", "RV", "PV", "RH", "PH", "VH",
                    "RR", "LR", "PR", "VR", "PL", "RL", "LL"]

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∏–∑–≤–ª–µ—á–µ–Ω–∏—è –±—É–∫–≤–µ–Ω–Ω–æ–π —á–∞—Å—Ç–∏ (–±–µ–∑ –Ω–æ–º–µ—Ä–∞)
def extract_pattern(filename):
    match = re.search(r"\d+\.(\w+)\.png", filename)  # –ò—â–µ–º "—á–∏—Å–ª–æ.–Ω–∞–∑–≤–∞–Ω–∏–µ.png"
    return match.group(1) if match else None  # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –±—É–∫–≤–µ–Ω–Ω—É—é —á–∞—Å—Ç—å

# –ü–æ–ª—É—á–∞–µ–º —Å–ø–∏—Å–∫–∏ –≤—Å–µ—Ö —Ñ–∞–π–ª–æ–≤ –≤ –ø–∞–ø–∫–∞—Ö
fresh_files = glob.glob(os.path.join(fresh_folder, "*.png"))
spoiled_files = glob.glob(os.path.join(spoiled_folder, "*.png"))

# –°–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä–∏ { "LP": ["1.LP.png", "2.LP.png", ...] }
fresh_dict = {extract_pattern(f): f for f in fresh_files if extract_pattern(f)}
spoiled_dict = {extract_pattern(f): f for f in spoiled_files if extract_pattern(f)}

# –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º —Ç–æ–ª—å–∫–æ –Ω–µ–¥–æ—Å—Ç–∞—é—â–∏–µ –∏–∑–º–µ—Ä–µ–Ω–∏—è
for pattern in missing_patterns:
    fresh_file = fresh_dict.get(pattern)
    spoiled_file = spoiled_dict.get(pattern)

    if not fresh_file or not spoiled_file:
        print(f"–ü—Ä–æ–ø—É—â–µ–Ω–æ: {pattern}, —Ç–∞–∫ –∫–∞–∫ –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç –æ–¥–∏–Ω –∏–∑ —Ñ–∞–π–ª–æ–≤!")
        continue

    output_file = os.path.join(half_fresh_folder, os.path.basename(fresh_file))

    # –ó–∞–≥—Ä—É–∂–∞–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
    fresh_img = np.array(Image.open(fresh_file), dtype=np.float32)
    spoiled_img = np.array(Image.open(spoiled_file), dtype=np.float32)

    # –í—ã—á–∏—Å–ª—è–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ
    half_fresh_img = (fresh_img + spoiled_img) / 2
    half_fresh_img = np.clip(half_fresh_img, 0, 255).astype(np.uint8)

    # –°–æ—Ö—Ä–∞–Ω—è–µ–º
    Image.fromarray(half_fresh_img).save(output_file)
    print(f"–°–æ—Ö—Ä–∞–Ω–µ–Ω–æ: {output_file}")
```
