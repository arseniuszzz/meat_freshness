```python
import shutil
import os

def zip_folder(folder_path):
    """–ê—Ä—Ö–∏–≤–∏—Ä—É–µ—Ç –ø–∞–ø–∫—É –≤ ZIP —Å —Ç–µ–º –∂–µ –∏–º–µ–Ω–µ–º (–±–µ–∑ /content/)"""
    folder_name = os.path.basename(folder_path)  # –ü–æ–ª—É—á–∞–µ–º –∏–º—è –ø–∞–ø–∫–∏ –±–µ–∑ –ø—É—Ç–∏
    zip_file_name = f"{folder_name}.zip"  # –î–æ–±–∞–≤–ª—è–µ–º .zip

    try:
        shutil.make_archive(zip_file_name[:-4], 'zip', folder_path)  # –°–æ–∑–¥–∞—ë–º –∞—Ä—Ö–∏–≤
        print(f"‚úÖ –£—Å–ø–µ—à–Ω–æ –∞—Ä—Ö–∏–≤–∏—Ä–æ–≤–∞–Ω–æ: {folder_name} ‚Üí {zip_file_name}")
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –∞—Ä—Ö–∏–≤–∞—Ü–∏–∏: {e}")

# üîπ –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ
folder_to_zip1 = "/content/SPOILT_BLUE_cropped2"  # –£–∫–∞–∂–∏ —Å–≤–æ—é –ø–∞–ø–∫—É
zip_folder(folder_to_zip1)
```

```python
# –ü—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞–ª—å–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏
data_dir = r'/content/SPOILT_BLUE_RAW'
# –ü–∞–ø–∫–∞ –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
output_dir = '/content/SPOILT_BLUE_cropped'

# –°–æ–∑–¥–∞–µ–º –ø–∞–ø–∫—É –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è, –µ—Å–ª–∏ –æ–Ω–∞ –µ—â–µ –Ω–µ —Å–æ–∑–¥–∞–Ω–∞
os.makedirs(output_dir, exist_ok=True)

# –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –≤—Å–µ–º –ø–∞–ø–∫–∞–º –≤–Ω—É—Ç—Ä–∏ –≥–ª–∞–≤–Ω–æ–π –ø–∞–ø–∫–∏
for folder_name in os.listdir(data_dir):
    folder_path = os.path.join(data_dir, folder_name)

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ —ç—Ç–æ –ø–∞–ø–∫–∞
    if os.path.isdir(folder_path):
        # –ù–∞—Ö–æ–¥–∏–º —Ñ–∞–π–ª —Å –æ–∫–æ–Ω—á–∞–Ω–∏–µ–º '_g.txt'
        for file_name in os.listdir(folder_path):
            if file_name.endswith('_r.txt'):
                file_path = os.path.join(folder_path, file_name)

                # –ß–∏—Ç–∞–µ–º –º–∞—Ç—Ä–∏—Ü—É —è—Ä–∫–æ—Å—Ç–∏ –∏–∑ —Ñ–∞–π–ª–∞
                brightness_matrix = np.loadtxt(file_path)

                # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å–ª–∏ –º–∞–∫—Å–∏–º—É–º –º–µ–Ω—å—à–µ 255, –Ω–æ—Ä–º–∞–ª–∏–∑—É–µ–º
                if brightness_matrix.max() < 255:
                    brightness_matrix = (brightness_matrix - brightness_matrix.min()) / \
                                        (brightness_matrix.max() - brightness_matrix.min()) * 255

                # –ü—Ä–∏–≤–æ–¥–∏–º –º–∞—Ç—Ä–∏—Ü—É –∫ uint8
                brightness_matrix = brightness_matrix.astype(np.uint8)

                # –°–æ–∑–¥–∞–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –∏ —Å–æ—Ö—Ä–∞–Ω—è–µ–º –µ–≥–æ –≤ PNG —Ñ–æ—Ä–º–∞—Ç–µ
                img = Image.fromarray(brightness_matrix)
                output_path = os.path.join(output_dir, f'{folder_name}.png')
                img.save(output_path)

                print(f'Saved {output_path}')
```

```python
import os
import numpy as np
from PIL import Image

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
threshold = 230 # –ü–æ—Ä–æ–≥ –ø–æ —è—Ä–∫–æ—Å—Ç–∏ –¥–ª—è –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è –ø—è—Ç–Ω–∞
output_size = 250  # –†–∞–∑–º–µ—Ä –≤—ã—Ä–µ–∑–∞–µ–º–æ–π –æ–±–ª–∞—Å—Ç–∏ (250x250)

# –ü—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å –¥–∞–Ω–Ω—ã–º–∏
data_dir = r'C:\Users\arsen\YandexDisk\–ö–æ–º–ø—å—é—Ç–µ—Ä DESKTOP-33KDD3Q\Neuro\–ò–∑–º–µ—Ä–µ–Ω–∏—è 11_10 —Å–≤–µ–∂–µ–µ –∫—Ä–∞—Å–Ω—ã–π'
# –ü–∞–ø–∫–∞ –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è –æ–±—Ä–µ–∑–∞–Ω–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
output_dir_cropped = 'FRESH_RED_cropped'

# –°–æ–∑–¥–∞–µ–º –ø–∞–ø–∫—É –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è, –µ—Å–ª–∏ –æ–Ω–∞ –µ—â–µ –Ω–µ —Å–æ–∑–¥–∞–Ω–∞
os.makedirs(output_dir_cropped, exist_ok=True)

# –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –≤—Å–µ–º –ø–∞–ø–∫–∞–º –≤–Ω—É—Ç—Ä–∏ –≥–ª–∞–≤–Ω–æ–π –ø–∞–ø–∫–∏
for folder_name in os.listdir(data_dir):
    folder_path = os.path.join(data_dir, folder_name)

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ —ç—Ç–æ –ø–∞–ø–∫–∞
    if os.path.isdir(folder_path):
        for file_name in os.listdir(folder_path):
            if file_name.endswith('_r.txt'):
                file_path = os.path.join(folder_path, file_name)

                # –ó–∞–≥—Ä—É–∂–∞–µ–º —è—Ä–∫–æ—Å—Ç—å –∏–∑ —Ñ–∞–π–ª–∞
                brightness_matrix = np.loadtxt(file_path)

                # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –Ω—É–∂–Ω–æ –ª–∏ –Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞—Ç—å
                if brightness_matrix.max() < 255:
                    brightness_matrix = (brightness_matrix - brightness_matrix.min()) / \
                                        (brightness_matrix.max() - brightness_matrix.min()) * 255
                brightness_matrix = brightness_matrix.astype(np.uint8)

                # –ü—Ä–∏–º–µ–Ω—è–µ–º –ø–æ—Ä–æ–≥: –æ—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ –ø–∏–∫—Å–µ–ª–∏ –≤—ã—à–µ –ø–æ—Ä–æ–≥–∞
                mask = brightness_matrix > threshold

                # –ü–æ–ª—É—á–∞–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –ø—è—Ç–Ω–∞ (bounding box)
                if np.any(mask):
                    coords = np.argwhere(mask)
                    y_min, x_min = coords.min(axis=0)
                    y_max, x_max = coords.max(axis=0)

                    # –ù–∞—Ö–æ–¥–∏–º —Ü–µ–Ω—Ç—Ä –ø—è—Ç–Ω–∞
                    center_y = (y_min + y_max) // 2
                    center_x = (x_min + x_max) // 2

                    # –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –¥–ª—è –≤—ã—Ä–µ–∑–∫–∏ 250x250
                    half_size = output_size // 2
                    top = max(center_y - half_size, 0)
                    left = max(center_x - half_size, 0)
                    bottom = min(center_y + half_size, brightness_matrix.shape[0])
                    right = min(center_x + half_size, brightness_matrix.shape[1])

                    # –û–±—Ä–µ–∑–∞–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –ø–æ —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω—ã–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º
                    cropped_matrix = brightness_matrix[top:bottom, left:right]

                    # –ï—Å–ª–∏ –≤—ã—Ä–µ–∑ –º–µ–Ω—å—à–µ 250x250, –¥–æ–ø–æ–ª–Ω—è–µ–º –µ–≥–æ –¥–æ –Ω—É–∂–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞
                    cropped_image = Image.fromarray(cropped_matrix)
                    cropped_image = cropped_image.resize((output_size, output_size), Image.Resampling.NEAREST)

                    # –°–æ—Ö—Ä–∞–Ω—è–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ
                    output_path_cropped = os.path.join(output_dir_cropped, f'{folder_name}.png')
                    cropped_image.save(output_path_cropped)

                    print(f'Saved cropped image: {output_path_cropped}')
```

```python
import os
import numpy as np
from PIL import Image
from scipy.ndimage import center_of_mass  # –ò–º–ø–æ—Ä—Ç —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Ü–µ–Ω—Ç—Ä–∞ –º–∞—Å—Å

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
threshold = 100  # –ü–æ—Ä–æ–≥ –ø–æ —è—Ä–∫–æ—Å—Ç–∏ –¥–ª—è –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è –ø—è—Ç–Ω–∞
output_size = 250  # –†–∞–∑–º–µ—Ä –≤—ã—Ä–µ–∑–∞–µ–º–æ–π –æ–±–ª–∞—Å—Ç–∏ (250x250)

# –ü—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å –¥–∞–Ω–Ω—ã–º–∏
data_dir = r'/content/SPOILT_BLUE_RAW'
# –ü–∞–ø–∫–∞ –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è –æ–±—Ä–µ–∑–∞–Ω–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
output_dir_cropped = '/content/SPOILT_BLUE_cropped2'

# –°–æ–∑–¥–∞–µ–º –ø–∞–ø–∫—É –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è, –µ—Å–ª–∏ –æ–Ω–∞ –µ—â–µ –Ω–µ —Å–æ–∑–¥–∞–Ω–∞
os.makedirs(output_dir_cropped, exist_ok=True)

# –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –≤—Å–µ–º –ø–∞–ø–∫–∞–º –≤–Ω—É—Ç—Ä–∏ –≥–ª–∞–≤–Ω–æ–π –ø–∞–ø–∫–∏
for folder_name in os.listdir(data_dir):
    folder_path = os.path.join(data_dir, folder_name)

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ —ç—Ç–æ –ø–∞–ø–∫–∞
    if os.path.isdir(folder_path):
        for file_name in os.listdir(folder_path):
            if file_name.endswith('_g.txt'):
                file_path = os.path.join(folder_path, file_name)

                # –ó–∞–≥—Ä—É–∂–∞–µ–º —è—Ä–∫–æ—Å—Ç—å –∏–∑ —Ñ–∞–π–ª–∞
                brightness_matrix = np.loadtxt(file_path)

                # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –Ω—É–∂–Ω–æ –ª–∏ –Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞—Ç—å
                if brightness_matrix.max() < 255:
                    brightness_matrix = (brightness_matrix - brightness_matrix.min()) / \
                                        (brightness_matrix.max() - brightness_matrix.min()) * 255
                brightness_matrix = brightness_matrix.astype(np.uint8)

                # –ü—Ä–∏–º–µ–Ω—è–µ–º –ø–æ—Ä–æ–≥: –æ—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ –ø–∏–∫—Å–µ–ª–∏ –≤—ã—à–µ –ø–æ—Ä–æ–≥–∞
                mask = brightness_matrix > threshold

                # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ —è—Ä–∫–æ–µ –ø—è—Ç–Ω–æ –Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–∏
                if np.any(mask):
                    # –í—ã—á–∏—Å–ª—è–µ–º —Ü–µ–Ω—Ç—Ä –º–∞—Å—Å —è—Ä–∫–∏—Ö –ø–∏–∫—Å–µ–ª–µ–π
                    center_y, center_x = center_of_mass(mask)
                    center_y, center_x = int(center_y), int(center_x)  # –ü—Ä–∏–≤–æ–¥–∏–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –∫ —Ü–µ–ª—ã–º —á–∏—Å–ª–∞–º

                    # –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –¥–ª—è –≤—ã—Ä–µ–∑–∫–∏ 250x250
                    half_size = output_size // 2
                    top = max(center_y - half_size, 0)
                    left = max(center_x - half_size, 0)
                    bottom = min(center_y + half_size, brightness_matrix.shape[0])
                    right = min(center_x + half_size, brightness_matrix.shape[1])

                    # –û–±—Ä–µ–∑–∞–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –ø–æ —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω—ã–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º
                    cropped_matrix = brightness_matrix[top:bottom, left:right]

                    # –ï—Å–ª–∏ –≤—ã—Ä–µ–∑ –º–µ–Ω—å—à–µ 250x250, –¥–æ–ø–æ–ª–Ω—è–µ–º –µ–≥–æ –¥–æ –Ω—É–∂–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞
                    cropped_image = Image.fromarray(cropped_matrix)
                    cropped_image = cropped_image.resize((output_size, output_size), Image.Resampling.NEAREST)

                    # –°–æ—Ö—Ä–∞–Ω—è–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ
                    output_path_cropped = os.path.join(output_dir_cropped, f'{folder_name}.png')
                    cropped_image.save(output_path_cropped)

                    print(f'Saved cropped image: {output_path_cropped}')
```

```python
# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞—Ç–∞—Å–µ—Ç–∞ –Ω–∞ 3 –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
import os
import shutil

# –ü—É—Ç—å –∫ –≥–ª–∞–≤–Ω–æ–π –ø–∞–ø–∫–µ, –≥–¥–µ –Ω–∞—Ö–æ–¥—è—Ç—Å—è –≤—Å–µ –ø–∞–ø–∫–∏ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏
source_dir = r'C:\Users\arsen\YandexDisk\–ö–æ–º–ø—å—é—Ç–µ—Ä DESKTOP-33KDD3Q\Neuro\Dataset'

# –ü–∞–ø–∫–∏ –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏
output_dirs = {
    'FRESH': os.path.join(source_dir, 'FRESH'),
    'HALF': os.path.join(source_dir, 'HALF'),
    'SPOILT': os.path.join(source_dir, 'SPOILT')
}

# –°–æ–∑–¥–∞–µ–º –ø–∞–ø–∫–∏ –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è, –µ—Å–ª–∏ –æ–Ω–∏ –µ—â–µ –Ω–µ —Å–æ–∑–¥–∞–Ω—ã
for path in output_dirs.values():
    os.makedirs(path, exist_ok=True)

# –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –ø–∞–ø–∫–∞–º –≤ source_dir
for folder_name in os.listdir(source_dir):
    folder_path = os.path.join(source_dir, folder_name)

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ —ç—Ç–æ –ø–∞–ø–∫–∞
    if os.path.isdir(folder_path):
        # –û–ø—Ä–µ–¥–µ–ª—è–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏—é (FRESH, HALF, SPOILT) –ø–æ –Ω–∞–∑–≤–∞–Ω–∏—é –ø–∞–ø–∫–∏
        if folder_name.startswith('FRESH'):
            category = 'FRESH'
        elif folder_name.startswith('HALF'):
            category = 'HALF'
        elif folder_name.startswith('SPOILT'):
            category = 'SPOILT'
        else:
            continue  # –ü—Ä–æ–ø—É—Å–∫–∞–µ–º –ø–∞–ø–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –æ—Ç–Ω–æ—Å—è—Ç—Å—è –∫ –Ω—É–∂–Ω—ã–º –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º

        # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Ñ–∞–π–ª–∞–º –≤ —Ç–µ–∫—É—â–µ–π –ø–∞–ø–∫–µ
        for file_name in os.listdir(folder_path):
            # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ —ç—Ç–æ —Ñ–∞–π–ª –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è (–Ω–∞–ø—Ä–∏–º–µ—Ä, PNG)
            if file_name.endswith('.png'):
                # –§–æ—Ä–º–∏—Ä—É–µ–º –Ω–æ–≤–æ–µ –∏–º—è —Ñ–∞–π–ª–∞, –¥–æ–±–∞–≤–∏–≤ –∫ –Ω–µ–º—É –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–∞–ø–∫–∏
                new_file_name = f"{folder_name}_{file_name}"

                # –ü—É—Ç—å –∫ –Ω–æ–≤–æ–º—É —Ñ–∞–π–ª—É
                new_file_path = os.path.join(output_dirs[category], new_file_name)

                # –ö–æ–ø–∏—Ä—É–µ–º —Ñ–∞–π–ª –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â—É—é –∫–∞—Ç–µ–≥–æ—Ä–∏—é
                shutil.copy2(os.path.join(folder_path, file_name), new_file_path)

                print(f"Copied {file_name} to {new_file_path}")
```
